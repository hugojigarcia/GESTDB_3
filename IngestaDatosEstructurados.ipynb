# Creacion del listado de URL para APIREST

```python
urlBase="https://api.jolpi.ca/ergast/f1"
```

```python
def get_data_from_api(endpoint, table_key_1, table_key_2):
    try:
        response = requests.get(f"{urlBase}/{endpoint}")
        # Verifica si la solicitud fue exitosa (código de estado 200)
        if response.status_code == 200:
            # Convierte la respuesta JSON en un diccionario
            data = response.json()
            data = data['MRData']
            print(data)
            # Extrae los datos de la clave específica (e.g., 'DriverTable', 'RaceTable')
            if table_key_1 in data:
                table_data = data[table_key_1][table_key_2]
                # Convierte la tabla en un DataFrame de pandas
                df = pd.DataFrame(table_data)
                print(f"Datos obtenidos de {endpoint} exitosamente.")
                return df
            else:
                print(f"La clave '{table_key_1}' no se encontró en la respuesta.")
                return pd.DataFrame()  # Retorna un DataFrame vacío si no se encuentra la clave
        else:
            print(f"Error {response.status_code} al obtener datos de {endpoint}.")
            return pd.DataFrame()  # Retorna un DataFrame vacío en caso de error
    except Exception as e:
        print(f"Error al conectar con la API: {e}")
        return pd.DataFrame()
```

```python
df_race = pd.read_csv("data/structured/races.csv")
df_drivers = get_data_from_api("drivers", "DriverTable", "Drivers")
# lee el csv data/structured/lap_times.csv
df_laps = pd.read_csv('data/structured/lap_times.csv')
# Muestra un resumen de los DataFrames obtenidos
print("Resumen de la tabla 'race':")
display(df_race)

print("\nResumen de la tabla 'drivers':")
display(df_drivers)

print("\nResumen de la tabla 'laps':")
display(df_laps)
```

# Creación de tablas

```python

cursor = conn.cursor()

# Consultas para eliminar las tablas si existen
drop_drivers_table = "DROP TABLE IF EXISTS Formula1_Drivers;"
drop_laps_table = "DROP TABLE IF EXISTS Formula1_Laps;"
drop_races_table = "DROP TABLE IF EXISTS Formula1_Races;"

# Ejecutar las consultas para eliminar las tablas
cursor.execute(drop_drivers_table)
cursor.execute(drop_laps_table)
cursor.execute(drop_races_table)

# Consultas para crear las tablas
create_drivers_table = """
CREATE TABLE IF NOT EXISTS Formula1_Drivers (
    driverId TEXT PRIMARY KEY,
    code TEXT,
    givenName TEXT NOT NULL,
    familyName TEXT NOT NULL,
    dateOfBirth DATE NOT NULL,
    nationality TEXT NOT NULL,
    url TEXT
);
"""

create_laps_table = """
CREATE TABLE IF NOT EXISTS Formula1_Laps (
    raceId INTEGER,
    driverId TEXT,
    lap INTEGER,
    position INTEGER,
    time TEXT,
    milliseconds INTEGER,
    PRIMARY KEY (raceId, driverId, lap)
);
"""

create_races_table = """
CREATE TABLE IF NOT EXISTS Formula1_Races (
    raceId INTEGER PRIMARY KEY,
    year INTEGER NOT NULL,
    round INTEGER NOT NULL,
    circuitId TEXT NOT NULL,
    name TEXT NOT NULL,
    date DATE NOT NULL,
    time TIME,
    url TEXT,
    fp1_date DATE,
    fp1_time TIME,
    fp2_date DATE,
    fp2_time TIME,
    fp3_date DATE,
    fp3_time TIME,
    quali_date DATE,
    quali_time TIME,
    sprint_date DATE,
    sprint_time TIME
);
"""

# Ejecutar las consultas para crear las tablas
cursor.execute(create_drivers_table)
cursor.execute(create_laps_table)
cursor.execute(create_races_table)

# Confirmar la transacción
conn.commit()

# Cerrar el cursor y la conexión
cursor.close()

print("Las tablas se han creado exitosamente.")
```

# Inserción de datos

```python
cursor = conn.cursor()

for index, row in df_race.iterrows():
    cursor.execute("""
        INSERT INTO Formula1_Races (
            raceId, year, round, circuitId, name, date, time, url,
            fp1_date, fp1_time, fp2_date, fp2_time,
            fp3_date, fp3_time, quali_date, quali_time,
            sprint_date, sprint_time
        ) VALUES (
            %s, %s, %s, %s, %s, %s, %s, %s,
            %s, %s, %s, %s,
            %s, %s, %s, %s,
            %s, %s
        )
        ON CONFLICT (raceId) DO UPDATE SET
            year = EXCLUDED.year,
            round = EXCLUDED.round,
            circuitId = EXCLUDED.circuitId,
            name = EXCLUDED.name,
            date = EXCLUDED.date,
            time = EXCLUDED.time,
            url = EXCLUDED.url,
            fp1_date = EXCLUDED.fp1_date,
            fp1_time = EXCLUDED.fp1_time,
            fp2_date = EXCLUDED.fp2_date,
            fp2_time = EXCLUDED.fp2_time,
            fp3_date = EXCLUDED.fp3_date,
            fp3_time = EXCLUDED.fp3_time,
            quali_date = EXCLUDED.quali_date,
            quali_time = EXCLUDED.quali_time,
            sprint_date = EXCLUDED.sprint_date,
            sprint_time = EXCLUDED.sprint_time;
    """, (
       row['raceId'], row['year'], row['round'], row['circuitId'], row['name'], 
        row['date'], row.get('time'), row.get('url'), row.get('fp1_date'), 
        row.get('fp1_time'), row.get('fp2_date'), row.get('fp2_time'),
        row.get('fp3_date'), row.get('fp3_time'), row.get('quali_date'),
        row.get('quali_time'), row.get('sprint_date'), row.get('sprint_time')
    ))
conn.commit()

    
for index, row in df_drivers.iterrows():
    cursor.execute("""
        INSERT INTO Formula1_Drivers (
            driverId, code, givenName, familyName, dateOfBirth, nationality, url
        ) VALUES (
            %s, %s, %s, %s, %s, %s, %s
        )
        ON CONFLICT (driverId) DO UPDATE SET
            code = EXCLUDED.code,
            givenName = EXCLUDED.givenName,
            familyName = EXCLUDED.familyName,
            dateOfBirth = EXCLUDED.dateOfBirth,
            nationality = EXCLUDED.nationality,
            url = EXCLUDED.url;
    """, (
        row['driverId'], row.get('code'),
        row['givenName'], row['familyName'], row['dateOfBirth'], row['nationality'], row.get('url')
    ))

# Confirmar la transacción
conn.commit()

for index, row in df_laps.iterrows():
    cursor.execute("""
        INSERT INTO Formula1_Laps (
            raceId, driverId, lap, position, time, milliseconds
        ) VALUES (
            %s, %s, %s, %s, %s, %s
        )
        ON CONFLICT (raceId, driverId, lap) DO UPDATE SET
            position = EXCLUDED.position,
            time = EXCLUDED.time,
            milliseconds = EXCLUDED.milliseconds;
    """, (
        row['raceId'], row['driverId'], row['lap'], row.get('position'),
        row.get('time'), row.get('milliseconds')
    ))

# Confirmar la transacción
conn.commit()


# Cerrar el cursor y la conexión
cursor.close()
conn.close()
```

# DUMP

```python
import psycopg2
import pandas as pd

# Conexión a la base de datos

# Crear un cursor para ejecutar comandos SQL
cursor = conn.cursor()

# Función para exportar una tabla a CSV
def export_table_to_csv(table_name, output_file):
    query = f"SELECT * FROM {table_name};"
    cursor.execute(query)
    rows = cursor.fetchall()
    columns = [desc[0] for desc in cursor.description]  # Obtener nombres de columnas

    # Crear un DataFrame de pandas
    df = pd.DataFrame(rows, columns=columns)

    # Exportar a CSV
    df.to_csv(output_file, index=False)

# Exportar las tablas deseadas
export_table_to_csv("Formula1_Drivers", "Formula1_Drivers.csv")
export_table_to_csv("Formula1_Laps", "Formula1_Laps.csv")
export_table_to_csv("Formula1_Races", "Formula1_Races.csv")

# Cerrar cursor y conexión
cursor.close()
conn.close()

print("Exportación de tablas completada.")

```
